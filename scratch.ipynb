```python
import numpy as np
import numpy as np

# Define the dimensions of your 2D array
rows, cols = 5, 5

# Generate a random 2D array
random_array = np.random.random((rows, cols))

# Create your original 2D array with 1s and 0s (e.g., using np.random.randint)
original_array = np.random.randint(0, 2, (rows, cols))
#original_array = np.zeros((rows, cols))

# Define a threshold value for the random numbers
threshold = 0.05

# Create a mask for values that were originally 1 or are neighbors of 1
mask = (original_array == 1)

# Check the conditions and update the random array
new_array = np.where(
    mask |\
        ((np.roll(mask, 1, axis=0)) |\
        (np.roll(mask, -1, axis=0)) |\
        (np.roll(mask, 1, axis=1)) |\
        (np.roll(mask, -1, axis=1))) &\
          (random_array < threshold), 
    1, 0)

# Print the original array
print("Original Array:")
print(original_array)

# Print the random array
print("Random Array:")
print(random_array)

# Print the new array based on your conditions
print("New Array:")
print(new_array)
```
